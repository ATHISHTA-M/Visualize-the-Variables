```python
import pandas as pd
import matplotlib.pyplot as plt
import zipfile
```

```python
file_path='/content/data.zip'
csv_file_name = 'API_SL.TLF.ACTI.FE.ZS_DS2_en_csv_v2_4978.csv'
with zipfile.ZipFile(file_path, 'r') as zip_ref:
  with zip_ref.open(csv_file_name) as csv_file:
    data = pd.read_csv(csv_file, skiprows=4)
```

```python
print(data.head())
```

```python
print(data.columns)
```

```python
data['Indicator Name'].value_counts().plot(kind='bar')
```

```python
year_column = '2022'
```

```python
plt.hist(data[year_column].dropna(), bins=10)
plt.title(f'Histogram of Labor Force Participation Rate ({year_column})')
plt.xlabel('Labor Force Participation Rate')
plt.ylabel('Frequency')
plt.show()
```
